```python
# Q 1. What are the five key concepts of Object-Oriented Programming (OOP)?

# ANS :

# 1. Encapsulation: Bundling data (attributes) and methods (functions) that operate on that data within a class.
#    It restricts direct access to some of an object's components, preventing accidental modification and ensuring data integrity.

# 2. Abstraction: Hiding complex implementation details and exposing only the necessary information to the user.
#    It simplifies interaction with objects and reduces complexity.

# 3. Inheritance: Creating new classes (derived classes) from existing ones (base classes).
#    It promotes code reuse and establishes a hierarchical relationship between classes, allowing for specialization and extension.

# 4. Polymorphism: The ability of objects of different classes to respond to the same method call in a way that is specific to their class.
#    It enables flexibility and allows for treating objects of different classes uniformly.

# 5. Association:  Describes relationships between different classes or objects.
#    It can be one-to-one, one-to-many, many-to-many, etc., and represents how objects interact with each other in a system.
```

```python
# Q 7. What is the Method Resolution Order (MRO) in Python? How can you retrieve it programmatically?


# ANS:


# The Method Resolution Order (MRO) in Python determines the order in which methods are searched for in a class hierarchy when a method is called on an object.
# It is especially important in multiple inheritance scenarios, where a class may inherit from multiple base classes, and there might be methods with the same name in different base classes.

# The MRO ensures that methods are resolved in a consistent and predictable manner, avoiding ambiguity and enabling proper inheritance behavior.

# You can retrieve the MRO of a class programmatically using the `__mro__` attribute or the `mro()` method.


# Example:

# class A:
#    def my_method(self):
#       print("Method from A")

# class B:
#    def my_method(self):
#        print("Method from B")

# class C(A, B):
#    pass

# obj = C()
# obj.my_method()  # Output will be "Method from A" based on MRO

# print(C.__mro__)  # Print the MRO of class C
# (<class '__main__.C'>, <class '__main__.A'>, <class '__main__.B'>, <class 'object'>)

# print(C.mro())  # Another way to print the MRO
# [<class '__main__.C'>, <class '__main__.A'>, <class '__main__.B'>, <class 'object'>]

# The MRO is calculated using a specific algorithm to ensure that it follows the following principles:

# 1. Child classes always precede parent classes.
# 2. The order of parent classes in the inheritance list is preserved.
# 3. The algorithm avoids ambiguity by ensuring that all base classes are visited in a linear order.

# In the example above, the MRO of class C is (C, A, B, object). When `obj.my_method()` is called, it first checks if C has a `my_method()`. Since it doesn't, it goes to the next class in the MRO, which is A. A has the method, so it is called.
```

```python
# Q 13. Explain the concept of the Diamond Problem in multiple inheritance. How does Python resolve it?

# ANS:

# The Diamond Problem:

# The Diamond Problem is a common issue in multiple inheritance where a class inherits from two or more classes that have a common ancestor.
# This creates an ambiguity when a method is called that is defined in the common ancestor because the interpreter does not know which version of the method to use (the one from the left parent or the one from the right parent).

# Example:

# class A:
#    def method(self):
#        print("Method from A")

# class B(A):
#    def method(self):
#        print("Method from B")

# class C(A):
#    def method(self):
#        print("Method from C")

# class D(B, C):
#    pass


# obj = D()
# obj.method() #Which method will be called?

# The Diamond Problem happens because class D inherits from both B and C, and both B and C inherit from A.
# If we try to call the `method` on the `D` object, it's not clear whether to use the implementation from B or C.

# Python's Solution: Method Resolution Order (MRO)

# Python resolves this issue using the Method Resolution Order (MRO).
# The MRO defines the order in which methods are searched for in a class hierarchy.
# Python uses the C3 linearization algorithm to compute the MRO, which ensures that:

# 1. Child classes are visited before parent classes.
# 2. The order of parent classes in the inheritance list is respected.
# 3. All base classes are visited in a linear order, avoiding ambiguity.

# In the example, the MRO of class D would be [D, B, C, A, object].
# Therefore, when `obj.method()` is called:

# 1. It searches for `method()` in D. It's not found.
# 2. It searches in B. B has `method()`, so it's called.
#
# So the output would be "Method from B".

# Note that if D has its own `method` definition, that would be called first.
# If not, the search continues as described in the MRO.

# In summary, Python's MRO ensures that the inheritance hierarchy is traversed in a consistent and predictable order, avoiding ambiguity and enabling proper method resolution in multiple inheritance scenarios.
```
